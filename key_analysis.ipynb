```python
def count_camelot_key_by_key_and_by_mode(df):
    

    djayPro_keys = ['1A', '2A', '3A', '4A', '5A', '6A', '7A', '8A', '9A', '10A', '11A', '12A',
                    '1B', '2B', '3B', '4B', '5B', '6B', '7B', '8B', '9B', '10B', '11B', '12B']
        
   # nb_items = df.shape[0]
    df = df.groupby(['track_key', 'track_mode', "djayPro_key"]).size()
    df = df.reset_index()
    df.columns = ['track_key', 'track_mode', 'djayPro_key', 'count']
    df['key-mode'] = df['track_key'].astype("string") + " , " + df['track_mode'].astype("string")
   # df = df.sort_values(by='count', ascending=False)
    #df['prob'] = (df['count'] / nb_items) * 100
    #df['djayPro_key'] = camelot_key
    
    return df
   # plot = df.plot(kind='bar', legend=False)
```

```python
def get_count_by_key_and_mode(df, key_mode, djayPro_key):
    try:
        count = df[(df['key-mode'] == key_mode) & (df['djayPro_key'] == djayPro_key)]['count'].values[0]
        return int(count)
    except:
        return 0

get_count_by_key_and_mode(df, '0 , 0', '5A')
```
